```python
import pandas as pd
data = pd.read_pickle('corpus.pkl')
data
```

```python
text = data.transcript.loc['ali']
text
```

```python
from collections import defaultdict

def markov_chain(text):
    words = text.split(" ")
    m_dict = defaultdict(list)
    for cur_word, next_word in zip(words[0:-1],words[1:]):
        m_dict[cur_word].append(next_word)
    m_dict = dict(m_dict)
    return m_dict
dic = markov_chain(text)
dic
```

```python
import random
def generate_sentences(chain,count = 15):
    word1 = random.choice(list(chain.keys()))
    sentence = word1.capitalize()
    for i in range(count):
        word2 = random.choice(chain[word1])
        word1 = word2
        sentence +=' '+word2
    sentence +='.'
    return sentence
generate_sentences(dic)
```
